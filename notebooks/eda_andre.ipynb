```python
import pandas as pd

df = pd.read_parquet('../data/lending_club_case_train_dataset_cleaned.parquet')
```

```python
df = df.dropna(subset=['default'])
```

```python
df['default'] = df['default'].astype(int)
```

```python
df['time_of_banking'] = (df['issue_d'] - df['earliest_cr_line']).dt.days
```

```python
print(df['time_of_banking'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))

# Use hue to color based on the "default" column
sns.histplot(df, x='time_of_banking', hue='default', bins=30, kde=True, alpha=0.5,stat='probability', common_norm=False)

# Labels and title
plt.xlabel('Days Difference')
plt.ylabel('Frequency')
plt.title('Distribution of Date Differences (Day) by Default Status')

plt.show()
```
